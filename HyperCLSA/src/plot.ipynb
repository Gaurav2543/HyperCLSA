```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Data preparation
views = [
    "All 3 considered", 
    "miRNA + meth", 
    "mRNA + meth", 
    "mRNA + miRNA", 
    "mRNA only", 
    "miRNA only", 
    "meth only"
]

acc = [0.90118, 0.81765, 0.89294, 0.88824, 0.88706, 0.78941, 0.79059]
f1_weighted = [0.90121, 0.81346, 0.89246, 0.88711, 0.88746, 0.78696, 0.78703]
f1_macro = [0.86572, 0.74645, 0.86134, 0.85533, 0.84437, 0.68984, 0.72331]

# Let's add some standard deviations to simulate error bars
std_acc = [0.02, 0.03, 0.025, 0.02, 0.03, 0.04, 0.035]
std_f1_weighted = [0.02, 0.03, 0.025, 0.02, 0.03, 0.04, 0.035]
std_f1_macro = [0.025, 0.035, 0.03, 0.025, 0.03, 0.045, 0.04]

# Set width of bars
barWidth = 0.25

# Set positions of the bars on X axis
r1 = np.arange(len(views))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

# Create the figure and axes
fig, ax = plt.subplots(figsize=(12, 6))

# Create bars with colors defined upfront
bars1 = ax.bar(r1, acc, width=barWidth, edgecolor='black', yerr=std_acc, capsize=5, label='Accuracy', color='#3498db')
bars2 = ax.bar(r2, f1_weighted, width=barWidth, edgecolor='black', yerr=std_f1_weighted, capsize=5, label='F1_weighted', color='#2ecc71')
bars3 = ax.bar(r3, f1_macro, width=barWidth, edgecolor='black', yerr=std_f1_macro, capsize=5, label='F1_macro', color='#e74c3c')

# Add title and axis names
ax.set_title('BRCA', fontsize=14, fontweight='bold')
ax.set_ylim(0.4, 1.0)  # Set y-axis limits similar to your image

# Add xticks on the middle of the group bars
ax.set_xticks([r + barWidth for r in range(len(views))])
ax.set_xticklabels(views, rotation=45, ha='right')

# Create legend & Show graphic
ax.legend(loc='upper right')

# Add grid lines
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Add value labels on bars
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=8)

# Uncomment if you want value labels on top of the bars
# add_labels(bars1)
# add_labels(bars2)
# add_labels(bars3)

plt.tight_layout()
# plt.savefig('brca_performance_comparison.png', dpi=300)
plt.show()
```
